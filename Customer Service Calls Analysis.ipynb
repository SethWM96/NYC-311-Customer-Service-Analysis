## Hypothesis:
1. Null Hypothesis (H0): There is no significatnt difference in closing time across locations.
2. Alternative Hypothesis (Ha): There is a significant difference in closing time across the cities. 

```python
import statsmodels.api as sm
from statsmodels.formula.api import ols
```

```python
model = ols('Request_Closing_Time ~ City', data=df).fit()
anova_table = sm.stats.anova_lm(model)
```

```python
p_value = anova_table['PR(>F)']['City']
print("ANOVA p-value:", p_value)
```

## Inferences from statistical analysis above:
1. p-value is 0.0
2. Null hypothesis rejected
3. There is a significant difference in closing times across cities

```python
#Kruskal Wallis Test
from scipy.stats import kruskal
```

```python
df_groups = df[['City','Request_Closing_Time']].dropna()

city_groups = {}
for city in df_groups['City'].unique():
    city_groups[city] = df_groups[df_groups['City'] == city]['Request_Closing_Time']
    
statistic, p_value = kruskal(*city_groups.values())

# Compare the p-value to the significance level
alpha = 0.05
if p_value > alpha:
    print("Fail to reject H0: All sample distributions are equal")
else:
    print("Reject H0: One or more sample distributions are not equal")
```

## Inferences from Kruskall Wallis Test:
1. Kruskall Wallis Test also confirms that we should reject H0
2. Cities have varying request closing times
